```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
import time
import numpy as np
```

```python
url = "https://fds.career.vt.edu/Salary/Details?cohort=2019-2020&dispmajor=Y"
resp= requests.get(url)
soup = bs(resp.text)
```

```python
data = []
for row in soup.select('tr'):
    row_text = [x.text for x in row.find_all('td')]
    data.append(row_text)
    print(', '.join(row_text))
```

```python
column_names = ["College/Major", "Graduates","Respondents","% Respondents", "Employed", "% Employed", "Reported Salary", "% Reported Salary", "25th Percentile", "Median", "75th Percentile"]
df = pd.DataFrame(data,columns = column_names )
df = df[1:].replace({r'\s+$': '', r'^\s+': ''}, regex=True).replace(r'\n',  ' ', regex=True)
df
```

```python
df.to_csv('salary_2019-2020.csv')
```
